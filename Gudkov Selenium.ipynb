```python
import numpy as np
import pandas as pd
import time

from itertools import compress
from selenium import webdriver as wb
```

```python
from selenium.webdriver.common.keys import Keys # for Enter or Return buttons, .send_keys(Keys.ENTER)
```

```python
data = pd.read_csv('candidates_table.csv')

disctricts = np.unique(data['район'])
```

```python
data['our_candidate'] = 0 # create a default column
```

```python
br = wb.Chrome("/Users/herrhellana/Downloads/chromedriver")
```

```python
br.get("https://mundep.gudkov.ru/")

find_button = br.find_element_by_name("address_line")
```

```python
# Looping through districts
for x in range(108, len(disctricts)):
    
    #### navigating ####
    find_button.clear() # clear from previous input
    time.sleep(3)
    find_button.send_keys(disctricts[x]) # type a district
    time.sleep(3)
    br.find_element_by_class_name("easy-autocomplete-container").click() # press 'enter'
    time.sleep(2)
    
    #### scrapping ####
    surnames_raw = br.find_elements_by_class_name("result-in-region__surname")
    names_raw = br.find_elements_by_class_name("rir__name")

    surnames = [i.get_attribute("innerHTML").split('> ')[1] for i in surnames_raw]
    names = [i.get_attribute("innerHTML").split('<strong>')[1].split('</strong>')[0] for i in names_raw]
    
    
    #### create an array of indices for 'our candidates' ####
    indices = []
    for i in range(len(surnames_raw)):
        if "our-candidate" in surnames_raw[i].get_attribute("innerHTML"):
            indices.append(i)

    
    # create a dataframe-like list of names
    district_our_candidates = list(map(lambda x: ' '.join(x), 
                               list(set(zip(np.array(surnames)[np.array(indices)], # our candidates, surnames
                                            np.array(names)[np.array(indices)]))))) # our candidates, names
    
    #### filling the dataset column ####
    data.loc[data['фио'].isin(district_our_candidates), 'our_candidate'] = 1
```
